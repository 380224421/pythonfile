```python
from wordcloud import WordCloud
import wordcloud
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
# 解决读取图片报错
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
import os


def word_cloud(word):

    wc_mask = np.array(Image.open('jack.jpg'))
    wc = wordcloud.WordCloud(
        # 设置字体格式
        font_path='buzz_cloud_font.ttf',
        # 设置背景图
        mask=wc_mask,
        # 最多显示词数
        max_words=200,
        # 字体最大值
        max_font_size=320,
        # 整体背景色
        background_color='White',
        # 词云的边框大小
        contour_width=2,
        # 词云的边框颜色
        contour_color='skyblue',
        # 设置字体重复
        repeat=True,
        color_func=lambda *args, **kwargs: "blue")

    # 从字典生成词云
    wc.generate_from_frequencies(word)
    # 从背景图建立颜色方案
    image_colors = wordcloud.ImageColorGenerator(wc_mask)
    # 显示词云
    plt.figure(figsize=[10, 10], dpi=300)
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
```

```python
dic = {'白':14,
       '富':13,
       '美':7}
```

```python
word_cloud(dic)
```
